```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "OrderDate": ["2024-01-01","2024-01-02","2024-01-03","2024-01-04","2024-01-05",
                  "2024-01-06","2024-01-07","2024-01-08","2024-01-09","2024-01-10"],
    "Product": ["Laptop","Phone","Tablet","Laptop","Phone",
                "Tablet","Laptop","Phone","Tablet","Laptop"],
    "Quantity": [2,5,3,1,4,2,3,6,2,1],
    "Price": [50000,20000,15000,50000,20000,
              15000,50000,20000,15000,50000]
}

df = pd.DataFrame(data)
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
df["TotalAmount"] = df["Quantity"] * df["Price"]

df.head()
df.isnull().sum()
df.duplicated().sum()
df.describe()
```

```python
df.describe()
total_revenue = df["TotalAmount"].sum()
print("Total Revenue:", total_revenue)
product_sales = df.groupby("Product")["TotalAmount"].sum()
print(product_sales)
print("Best Product:", product_sales.idxmax())
print("Average Order Value:", df["TotalAmount"].mean())
```

```python
product_sales.plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
df.groupby("OrderDate")["TotalAmount"].sum().plot(kind="line")
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

```python
product_sales.plot(kind="pie", autopct='%1.1f%%')
plt.title("Revenue Distribution")
plt.ylabel("")
plt.show()
```

```python
print("Mean:", df["TotalAmount"].mean())
print("Standard Deviation:", df["TotalAmount"].std())
print("Correlation:", df["Quantity"].corr(df["TotalAmount"]))
```
