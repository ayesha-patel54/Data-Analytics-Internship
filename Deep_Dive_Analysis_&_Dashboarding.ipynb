```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    "OrderDate": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Product": ["Laptop", "Phone", "Tablet", "Laptop", "Phone"],
    "Quantity": [2, 5, 3, 1, 4],
    "Price": [50000, 20000, 15000, 50000, 20000]
}

df = pd.DataFrame(data)
df["TotalAmount"] = df["Quantity"] * df["Price"]
df["OrderDate"] = pd.to_datetime(df["OrderDate"])

df.head()
```

```python
total_revenue = df["TotalAmount"].sum()
print("Total Revenue:", total_revenue)
```

```python
product_sales = df.groupby("Product")["TotalAmount"].sum()
print(product_sales)
print("Top Product:", product_sales.idxmax())
```

```python
avg_order = df["TotalAmount"].mean()
print("Average Order Value:", avg_order)
```
