```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {
    "OrderDate": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Product": ["Laptop", "Phone", "Tablet", "Laptop", "Phone"],
    "Quantity": [2, 5, 3, 1, 4],
    "Price": [50000, 20000, 15000, 50000, 20000]
}

df = pd.DataFrame(data)
df["TotalAmount"] = df["Quantity"] * df["Price"]

df.head()
```

```python
print("Mean:", df["TotalAmount"].mean())
print("Standard Deviation:", df["TotalAmount"].std())
```

```python
correlation = df["Quantity"].corr(df["TotalAmount"])
print("Correlation:", correlation)
```

```python
laptop_mean = df[df["Product"]=="Laptop"]["TotalAmount"].mean()
phone_mean = df[df["Product"]=="Phone"]["TotalAmount"].mean()

print("Laptop Mean:", laptop_mean)
print("Phone Mean:", phone_mean)
```
